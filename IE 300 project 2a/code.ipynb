```python
import numpy as np
from scipy.stats import binom
from scipy.stats import poisson
from scipy.stats import norm
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import pandas as pd
#input values
file = "Isermann_Bluegill_sampling_catch_1.csv"
df = pd.read_csv(file)
clean_df = df.dropna(how = "any")
#statistical characteristics
mean = np.mean(clean_df["Bluegill Captured"])
variance = np.var(clean_df["Bluegill Captured"])
median = np.median(clean_df["Bluegill Captured"])
rangee = np.max(clean_df["Bluegill Captured"]) - np.min(clean_df["Bluegill Captured"])
print(mean)
print(variance)
print(median)
print(rangee)
data = clean_df["Bluegill Captured"].tolist()
#finding outliers using boxplots
Q1,Q3 = np.percentile(data,[25,75])
IQR = Q3-Q1
Lowerbound = Q1 - 1.5*IQR
Upperbound = Q3 + 1.5*IQR
outlier = []
for j in range (len(data)):
    if (data[j] < Lowerbound or data[j] > Upperbound):
        outlier.append(data[j])
print(outlier)
#generate frequency histogram
plt.hist(data,bins=25)
#normal approximation
x = np.linspace(min(data),max(data),1000)
y = norm(loc=mean,scale = np.sqrt(variance)).pdf(x)
plt.plot(x,100*y,label= "Normal Approximation")
plt.show()
```
